<p> We need to do the calculation of median of Experience and Salary</p>
<p> We need find out at least one point which belongs to the Regression line. </p>

$$\hat{x} ,  \hat{y}$$

Let's start by 

$$\hat{x} = \sum \limits _{j=1} ^n x \div len(x)$$  
or 
$$\hat{x} = \frac{x1+ x2 + x3....+ xn}{len(x)} $$

then 

$$\hat{y} = \sum \limits _{j=1} ^n y \div len(y)$$  
or 
$$\hat{y} = \frac{y1+ y2 + y3....+ yn}{len(y)} $$

```python
# Complete
x_hat = X.sum() / len(X)
x_hat
```

```python
# Complete
y_hat = sum(y)/len(y)

y_hat
```

###<div> Nous connaissons au moins un point de <strong><em>la droite de regression</em></strong>, représentons le ( forme : étoile )</div>

```python
plt.scatter(x=X,y=y)
plt.scatter(x_hat, y_hat, marker='*',c=['#ff7f0e'], s=200)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Salary base on Experience")
plt.show()
```

<div>Next we need to calculate the distance between those The median points and the real salary points.</div>
<div>Calculate the distance between these median point and other points.</div>

by doing:
En faisant:
$${x} - \hat{x} $$ 
and 
$${y} - \hat{y}$$

```python
# Complete

dif_x = X - x_hat
dif_x
```

```python
# Complete 

dif_y = y- y_hat
dif_y
```

$${x} - \hat{x} $$ 
and 

$${y} - \hat{y}$$

The slope : la pente

$$\hat{m} = \frac{ \sum(({x} - \hat{x}) ({y} - \hat{y}))}{\sum({x} - \hat{x})^2} $$ 
m est la pente( la bonne inclinaison de la droite) qui passe par le point.

```python
# Complete
m = sum(dif_x * dif_y)/ sum(dif_x**2)
m
```

$$\hat{y} = m\hat{x} + c$$

+ The slope is known => m
+ The y is known
+ The x is known

$$ c = \hat{y} - m\hat{x}$$

```python
# Complete

c = y_hat - m*x_hat
c
```

We will make a function using ⏰
+ m [slope]
+ c [intercept]:

```python
def get_linreg(arr): # arr ==> vector des experiences 
  return m*arr + c
```

```python
# Get all predicted salaries as a line

salary_predicted = get_linreg(X)
```

```python
salary_predicted
```

```python
salary_predicted - y
```
